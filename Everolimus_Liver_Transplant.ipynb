# Introduction

Liver transplantation remains the definitive treatment for end-stage liver disease and acute liver failure. Despite advances in surgical techniques, perioperative care, and immunosuppressive strategies, allograft rejection continues to represent a major cause of graft dysfunction and long-term morbidity following transplantation.

Conventional post-transplant immunosuppressive regimens commonly involve tacrolimus-based therapy in combination with corticosteroids and anti-proliferative agents such as mycophenolate mofetil. Although tacrolimus remains the cornerstone of liver transplant immunosuppression, prolonged calcineurin inhibitor exposure has been associated with nephrotoxicity, metabolic complications, and progressive renal dysfunction.

In selected patients with persistent rejection or incomplete biochemical response, escalation of immunosuppressive therapy may be required. Everolimus, a mammalian target of rapamycin (mTOR) inhibitor, has emerged as an adjunct immunosuppressive strategy capable of facilitating calcineurin inhibitor minimization while potentially contributing to graft stabilization and rejection control.

Current literature evaluating Everolimus in liver transplantation has largely focused on graft survival, rejection prevention, and renal preservation; however, longitudinal real-world biochemical response patterns following Everolimus-based escalation therapy remain incompletely characterized in heterogeneous liver allograft rejection populations.

Additionally, relatively few real-world analytical studies have explored longitudinal biomarker evolution using reproducible computational healthcare analytics methodologies within heterogeneous transplant rejection cohorts.

This project therefore performs a retrospective longitudinal analytical study evaluating biochemical trajectories, renal function trends, tacrolimus exposure, Everolimus timing, and clinical response patterns among liver transplant rejection patients receiving Everolimus-based escalation therapy using reproducible Python-based healthcare analytics workflows.

# Clinical Rationale

Management of liver allograft rejection remains clinically challenging, particularly among patients demonstrating incomplete biochemical or clinical response to conventional calcineurin inhibitor-based immunosuppression.

Standard rejection management commonly involves tacrolimus, corticosteroids, and mycophenolate mofetil-based therapy initiated following rejection confirmation. However, selected patients may demonstrate persistent graft dysfunction, worsening cholestatic injury, progressive liver enzyme elevation, or concerns regarding calcineurin inhibitor-related toxicity despite conventional treatment.

Everolimus, an mTOR inhibitor, has emerged as a potential Everolimus-based escalation therapy in such patients. In addition to its immunomodulatory effects, Everolimus may facilitate tacrolimus minimization while contributing to renal preservation and graft-related biochemical stabilization.

This study therefore aims to evaluate longitudinal biochemical, renal, and immunosuppressive response patterns following Everolimus-based escalation therapy using retrospective transplant healthcare analytics workflows.

```python
# ============================================
# FINAL ROBUST PT/INR EXTRACTION
# ============================================

import numpy as np

pt_inr_mapping = {

    'pre': ('PRE-EVEROLIMUS VALUES', 'PT/INR'),

    'day7': ('PT/INR', 'DAY 7'),

    'day30': ('PT/INR', 'DAY 30'),

    'day45': ('PT/INR', 'DAY 45'),

    'day60': ('PT/INR', 'DAY 60'),

    'day90': ('PT/INR', 'DAY 90'),

    'month6': ('PT/INR', 'MONTH 6'),

    'month12': ('PT/INR', 'MONTH 12'),

    'month18': ('PT/INR', 'MONTH 18'),

    'recent': (
        'Last Date of Visit/Recent Follow up Lab values',
        'PT/INR'
    )
}

for tp, col in pt_inr_mapping.items():

    # Load raw values safely

    values = raw_df[col].astype(str)

    # Fix malformed INR formatting

    values = values.str.replace(
        '/0/',
        '/0.',
        regex=False
    )

    pt_values = []
    inr_values = []

    for val in values:

        # Missing values

        if val in ['nan', 'NaN', '', 'None']:

            pt_values.append(np.nan)
            inr_values.append(np.nan)
            continue

        # Proper PT/INR format

        if '/' in val:

            split = val.split('/', 1)

            pt = pd.to_numeric(
                split[0],
                errors='coerce'
            )

            inr = pd.to_numeric(
                split[1],
                errors='coerce'
            )

            pt_values.append(pt)
            inr_values.append(inr)

        else:

            # PT only without INR

            pt = pd.to_numeric(
                val,
                errors='coerce'
            )

            pt_values.append(pt)
            inr_values.append(np.nan)

    analysis_df[f'pt_{tp}'] = pt_values

    analysis_df[f'inr_{tp}'] = inr_values
```

```python
# ============================================
# INR VALUE SANITY CORRECTION
# ============================================

for col in analysis_df.columns:

    if col.startswith('inr_'):

        analysis_df.loc[
            analysis_df[col] > 10,
            col
        ] = analysis_df.loc[
            analysis_df[col] > 10,
            col
        ] / 10
```

# Statistical Analysis

Given the small retrospective cohort size and anticipated non-normal biomarker distributions, non-parametric statistical methods were utilized for exploratory comparative analyses.

Paired longitudinal comparisons were evaluated using Wilcoxon signed-rank testing, while subgroup comparisons between responders and non-responders were performed using Mann–Whitney U testing.

Exploratory analyses were interpreted cautiously due to limited sample size and heterogeneity in longitudinal follow-up availability.

```python
# ============================================
# STATISTICAL LIBRARIES
# ============================================

from scipy.stats import wilcoxon

from scipy.stats import mannwhitneyu

from scipy.stats import spearmanr
```

## Paired Longitudinal Bilirubin Analysis

Paired non-parametric longitudinal analysis was performed to compare pre-Everolimus bilirubin levels with final follow-up bilirubin values following escalation therapy.

```python
# ============================================
# WILCOXON TEST:
# PRE VS FINAL BILIRUBIN
# ============================================

paired_bili = analysis_df[
    [
        'bilirubin_pre_everolimus',
        'harmonized_final_bilirubin'
    ]
].dropna()

wilcoxon_result = wilcoxon(

    paired_bili['bilirubin_pre_everolimus'],

    paired_bili['harmonized_final_bilirubin']

)

print(wilcoxon_result)
```

### Interpretation

Paired longitudinal analysis demonstrated a trend toward bilirubin improvement following Everolimus escalation therapy, although statistical significance was not reached within the current retrospective cohort (p = 0.089).

The absence of formal statistical significance likely reflects the limited sample size, heterogeneous longitudinal follow-up availability, and substantial inter-patient variability observed within the cohort.

Observed longitudinal trajectory patterns nevertheless suggested potential biochemical improvement following immunosuppressive optimization in selected patients..

## Comparative Timing Analysis Between Responders and Non-Responders

Non-parametric subgroup comparison analysis was performed to evaluate whether time-to-Everolimus initiation differed between responder and non-responder groups.

```python
# ============================================
# MANN-WHITNEY U TEST
# EVEROLIMUS TIMING
# ============================================

responders_timing = analysis_df[
    analysis_df['response_group'] == 'Responder'
][
    'days_rejection_to_everolimus'
].dropna()

nonresponders_timing = analysis_df[
    analysis_df['response_group'] == 'Non-Responder'
][
    'days_rejection_to_everolimus'
].dropna()

mannwhitney_result = mannwhitneyu(

    responders_timing,

    nonresponders_timing,

    alternative='two-sided'
)

print(mannwhitney_result)
```

### Interpretation

Exploratory subgroup comparison analysis was performed to evaluate potential differences in Everolimus initiation timing between responders and non-responders.

Observed trends may suggest possible associations between escalation timing and biochemical response patterns. However, findings should be interpreted cautiously given the limited cohort size and retrospective observational study design.

## Correlation Between Everolimus Initiation Timing and Final Bilirubin

Exploratory correlation analysis was performed to evaluate whether delayed Everolimus initiation demonstrated associations with final bilirubin outcomes following escalation therapy.

```python
# ============================================
# SPEARMAN CORRELATION
# ============================================

correlation_df = analysis_df[
    [
        'days_rejection_to_everolimus',
        'harmonized_final_bilirubin'
    ]
].dropna()

spearman_result = spearmanr(

    correlation_df['days_rejection_to_everolimus'],

    correlation_df['harmonized_final_bilirubin']
)

print(spearman_result)
```
